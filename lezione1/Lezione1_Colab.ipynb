## рҹҡҖ Step 2 вҖ” La prima chiamata all'API

Questo ГЁ il codice piГ№ importante del corso. Leggi ogni riga con attenzione prima di eseguire.

````python
# La prima chiamata a Claude
message = client.messages.create(
    model="claude-haiku-4-5-20251001",   # modello veloce per esercizi
    max_tokens=1024,
    messages=[
        {
            "role": "user",
            "content": "Ciao! Presentati in 2 righe e dimmi cosa puoi fare per me."
        }
    ]
)

print("=" * 50)
print("RISPOSTA DI CLAUDE:")
print("=" * 50)
print(message.content[0].text)
````

````python
# Esploriamo la struttura completa della risposta
print("рҹ“Ҡ Dettagli della risposta:")
print(f"  Modello usato:     {message.model}")
print(f"  Stop reason:       {message.stop_reason}")
print(f"  Token in input:    {message.usage.input_tokens}")
print(f"  Token in output:   {message.usage.output_tokens}")
print(f"  Token totali:      {message.usage.input_tokens + message.usage.output_tokens}")
````

## рҹҢЎпёҸ Step 3 вҖ” Il parametro `temperature`

La `temperature` controlla quanto ГЁ variabile la risposta:
- **0.0** вҶ’ sempre la stessa risposta (codice, calcoli, fatti)
- **0.7** вҶ’ equilibrio (default consigliato)
- **1.0** вҶ’ molto creativo e variabile (scrittura creativa)

````python
# Funzione helper вҖ” la useremo in tutti gli esercizi
def chiedi_claude(domanda, temperature=0.7, system=None, max_tokens=300):
    """Invia una domanda a Claude e restituisce la risposta testuale."""
    params = {
        "model": "claude-haiku-4-5-20251001",
        "max_tokens": max_tokens,
        "temperature": temperature,
        "messages": [{"role": "user", "content": domanda}]
    }
    if system:
        params["system"] = system

    response = client.messages.create(**params)
    return response.content[0].text

print("вң… Funzione helper pronta!")
````

````python
# Stessa domanda, temperature diverse
domanda = "Dammi un nome creativo per un chatbot AI per studenti universitari."

print("рҹҘ¶ temperature = 0.0 (deterministico):")
print(chiedi_claude(domanda, temperature=0.0))
print()

print("рҹҢЎпёҸ  temperature = 0.7 (bilanciato):")
print(chiedi_claude(domanda, temperature=0.7))
print()

print("рҹ”Ҙ temperature = 1.0 (creativo):")
print(chiedi_claude(domanda, temperature=1.0))
````

## рҹҺӯ Step 4 вҖ” Il `system` prompt

Il system prompt definisce la **personalitГ ** e il **comportamento** del chatbot.  
Stesso modello, system prompt diversi вҶ’ comportamenti completamente diversi.

````python
domanda = "Come funziona Python?"

print("рҹ‘Ө Senza system prompt:")
print(chiedi_claude(domanda))
print()

print("рҹ‘©вҖҚрҹҸ« Insegnante per bambini delle elementari:")
print(chiedi_claude(domanda,
    system="Sei un insegnante per bambini di 8 anni. Spiega tutto con analogie del mondo reale, senza termini tecnici."))
print()

print("рҹ‘ЁвҖҚрҹ’» Sviluppatore senior:")
print(chiedi_claude(domanda,
    system="Sei uno sviluppatore senior con 20 anni di esperienza. Rispondi in modo tecnico e diretto."))
````
